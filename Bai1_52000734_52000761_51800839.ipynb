```python
import pandas as pd
import numpy as np
from time import time
```

#Reading data

```python
df = pd.read_csv('bank-marketing-data.csv', sep = ';')
```

```python
from sklearn.preprocessing import LabelEncoder
class MultiColumnLabelEncoder:
    def __init__(self,columns = None):
        self.columns = columns # array of column names to encode

    def fit(self,X,y=None):
        return self # not relevant here

    def transform(self,X):
        '''
        Transforms columns of X specified in self.columns using
        LabelEncoder(). If no columns specified, transforms all
        columns in X.
        '''
        output = X.copy()
        if self.columns is not None:
            for col in self.columns:
                output[col] = LabelEncoder().fit_transform(output[col])
        else:
            for colname,col in output.iteritems():
                output[colname] = LabelEncoder().fit_transform(col)
        return output

    def fit_transform(self,X,y=None):
        return self.fit(X,y).transform(X)
```

```python
header = df.columns
```

```python
from collections import Counter
count = Counter(df[header[20]])
print(count)
```

```python
df.head(5)
```

#Preprocessing

```python
df = MultiColumnLabelEncoder(columns = [header[1],header[2], header[3], header[4], header[5], header[6], header[7], header[8], header[9], header[10], header[14]]).fit_transform(df)
df[header[20]] = LabelEncoder().fit_transform(df[header[20]])
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
def scaleColumns(df, cols_to_scale):
    for col in cols_to_scale:
        df[col] = pd.DataFrame(scaler.fit_transform(pd.DataFrame(df[col])),columns=[col])
    return df
scaled_df = scaleColumns(df, [header[0], header[11], header[12], header[13], header[15], header[16], header[17], header[18], header[19]])
scaled_df
```

```python
X = scaled_df.values[:,:-1]
y = scaled_df.values[:,-1]
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10) 
```

#Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB 
gnb = GaussianNB() 
start = time()
gnb.fit(X_train, y_train)
end = time() 
training_time_gnb = end - start

start = time()
y_pred_gnb = gnb.predict(X_test)
end = time() 
testing_time_gnb = end - start
```

#Weighted kNN (k = 7)

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7, p=1, weights = 'distance')
start = time()
knn.fit(X_train,y_train)
end = time() 
training_time_knn = end - start

start = time()
y_pred_knn = knn.predict(X_test)
end = time() 
testing_time_knn = end - start
```

#Decision Tree (max_depth = 5)

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth = 5)
start = time()
dt.fit(X_train, y_train)
end = time() 
training_time_dt = end - start

start = time()
y_pred_dt = dt.predict(X_test)
end = time() 
testing_time_dt = end - start
```

#Accuracy Score

```python
from sklearn import metrics 
print("Gaussian Naive Bayes model accuracy score:", metrics.accuracy_score(y_test, y_pred_gnb))
print("Gaussian kNN model accuracy score:", metrics.accuracy_score(y_test, y_pred_knn))
print("Gaussian Decision Tree model accuracy score:", metrics.accuracy_score(y_test, y_pred_dt))
```

#Precision, recall, f1-score và weighted average of f1-score của toàn bộ dữ liệu

```python
print("Report of Gaussian Naive Bayes:")
print(metrics.classification_report(y_test,y_pred_gnb))
```

```python
print("Report of kNN:")
print(metrics.classification_report(y_test,y_pred_knn))
```

```python
print("Report of Decision Tree:")
print(metrics.classification_report(y_test,y_pred_dt))
```

#Thời gian train của các models: Gaussian Naive Bayess, kNN và Decision Tree

```python
print("Gaussian Naive Bayes training time: ", training_time_gnb)
print("kNN training time: ", training_time_knn)
print("Decision training time: ", training_time_dt)
```

#Thời gian test của các models: Gaussian Naive Bayess, kNN và Decision Tree

```python
print("Gaussian Naive Bayes testing time: ", testing_time_gnb)
print("kNN testing time: ", testing_time_knn)
print("Decision testing time: ", testing_time_dt)
```
